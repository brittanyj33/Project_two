```python
#imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
#import alpaca_trade_api as tradeapi
#from MCForecastTools import MCSimulation
#import plotly.express as px
import pandas as pd
#import hvplot.pandas
import numpy as np
#import matplotlib.pyplot as plt
from pathlib import Path
#import holoviews as hv
#import json
from pandas_datareader import data as pdr
from datetime import date
import yfinance as yf
yf.pdr_override()
from yahoo_fin.stock_info import get_data

%matplotlib inline
```

```python
load_dotenv()
stock_tickers = ["AMZN","AAPL","NFLX"]
```

```python
amzn_data_2016 = get_data(stock_tickers[0], start_date = "2016/04/02", end_date = "2016/06/15", index_as_date = False, interval = "1d")
amzn_data_2016
```

```python
amzn_2016_final = amzn_data_2016.drop(["open", "high","low","close","volume"], axis=1)
amzn_2016_final
```

```python
amzn_data_2016_final = amzn_2016_final.pivot_table(values = "adjclose", index = "date", columns = "ticker")
amzn_data_2016_final
```

```python
aapl_data_2016 = get_data(stock_tickers[1], start_date = "2016/04/02", end_date = "2016/06/15", index_as_date = False, interval = "1d")
aapl_data_2016
```

```python
aapl_data_final = aapl_data_2016.drop(["open", "high","low","close","volume"], axis=1)
aapl_data_final
```

```python
aapl_data_2016_final = aapl_data_final.pivot_table(values = "adjclose", index = "date", columns = "ticker")
aapl_data_2016_final
```

```python
nflx_data_2016 = get_data(stock_tickers[2], start_date = "2016/04/02", end_date = "2016/06/15", index_as_date = False, interval = "1d")
nflx_data_2016
```

```python
nflx_data_final = nflx_data_2016.drop(["open", "high","low","close","volume"], axis=1)
nflx_data_final
```

```python
nflx_data_2016_final = nflx_data_2016.pivot_table(values = "adjclose", index = "date", columns = "ticker")
nflx_data_2016_final
```

```python
stock_data_2016_df = pd.concat([amzn_data_2016_final,aapl_data_2016_final,nflx_data_2016_final], axis = "columns", join = "inner")
stock_data_2016_df
```

```python
stock_data_2016_df.to_csv(r'../Project_two/Stock-Analysis/2016_stock_data.csv', index = True)
```
